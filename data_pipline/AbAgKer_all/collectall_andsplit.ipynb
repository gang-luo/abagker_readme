### randmom 5-fold

```python
#!/usr/bin/env python3
"""
五折交叉验证数据分割脚本：将JSON文件中的数据划分为5个fold，每个fold包含训练集和验证集
"""

import json
import random
import os
from typing import List, Dict, Any
import argparse


def save_json_data(data, file_path):
    
    os.makedirs(os.path.dirname(file_path), exist_ok=True)
    with open(file_path, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=2)
    print(f"success save: {file_path}")


def k_fold_split_dataset(file_path: str, output_dir: str, k: int = 5, random_seed: int = 42):
    """
    Args:
        file_path: json file
        output_dir:
        k: fold,5
        random_seed: seed
    """
    random.seed(random_seed)

    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)

    shuffled_data = data.copy()
    random.shuffle(shuffled_data)

    total_samples = len(shuffled_data)
    fold_size = total_samples // k
    remainder = total_samples % k  

    folds = []
    start_idx = 0
    for i in range(k):
        
        end_idx = start_idx + fold_size + (1 if i < remainder else 0)
        folds.append(shuffled_data[start_idx:end_idx])
        start_idx = end_idx

    base_name = os.path.splitext(os.path.basename(file_path))[0]

    
    for fold_idx in range(1, k + 1):
        print(f"\n process in {fold_idx}fold ...")

        val_data = folds[fold_idx - 1]
        train_data = [item for j, fold in enumerate(folds) if j != fold_idx - 1 for item in fold]

        train_file = os.path.join(output_dir, f"fold_{fold_idx}_train.json")
        val_file = os.path.join(output_dir, f"fold_{fold_idx}_test.json")

        save_json_data(train_data, train_file)
        save_json_data(val_data, val_file)

        print(f"traing nums: {len(train_data)}")
        print(f"val nums: {len(val_data)}")
        print(f"percent: {len(val_data)/total_samples:.3f}")


if __name__ == "__main__":
    
    random_seed = 42
    k_folds = 5
    file_path = "/root/private_data/luog/Data_AbAg/AbAgKer_all/json/AbAgI_9678.json"
    output_dir = "/root/private_data/luog/Data_AbAg/AbAgKer_all/split_fivefold/AbAgI"

    # file_path = "/root/private_data/luog/Data_AbAg/AbAgKer_all/json/AbAgA_1953.json"
    # output_dir = "/root/private_data/luog/Data_AbAg/AbAgKer_all/split_fivefold/AbAgA"

    # file_path = "/root/private_data/luog/Data_AbAg/AbAgKer_all/json/AbAgAoff_272.json"
    # output_dir = "/root/private_data/luog/Data_AbAg/AbAgKer_all/split_fivefold/AbAgIoff"


    k_fold_split_dataset(file_path, output_dir, k=k_folds, random_seed=random_seed)
```

## 5-fold based PDB name

```python
import json
from sklearn.model_selection import KFold
import random
from collections import defaultdict
import os

def split_pdb_data_for_cv(data, n_splits=5, random_seed=42):
    """
    using pdb top-4 cahr to condcut 5 fold
    :param data: list of dicts, 
    :param n_splits: int, fold, 5
    :param random_seed: seed
    :return: list of (train_data, val_data) tuples, 
    """
    random.seed(random_seed)
    # Step 1: /
    pdb_to_samples = defaultdict(list)
    for item in data:
        pdb_id = item["pdb"][:4]  # top-4
        pdb_to_samples[pdb_id].append(item)

    # Step 2: unique PDB 
    unique_pdbs = list(pdb_to_samples.keys())
    random.shuffle(unique_pdbs)  # random

    # Step 3: split fold
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=random_seed)
    folds = []

    for train_idx, val_idx in kf.split(unique_pdbs):
        train_pdbs = [unique_pdbs[i] for i in train_idx]
        val_pdbs = [unique_pdbs[i] for i in val_idx]

        train_data = []
        val_data = []

        for pdb in train_pdbs:
            train_data.extend(pdb_to_samples[pdb])
        for pdb in val_pdbs:
            val_data.extend(pdb_to_samples[pdb])

        folds.append((train_data, val_data))

    return folds


if __name__ == "__main__":
    
    output_dir = "/root/private_data/luog/Data_AbAg/AbAgKer_all/split_fivefold/AbAgI_byPDB"
    with open("/root/private_data/luog/Data_AbAg/AbAgKer_all/json/AbAgI_9678.json", "r") as f:
        data = json.load(f)

        
    # output_dir = "/root/private_data/luog/Data_AbAg/AbAgKer_all/split_fivefold/AbAgA_byPDB"
    # with open("/root/private_data/luog/Data_AbAg/AbAgKer_all/json/AbAgA_1953.json", "r") as f:
    #     data = json.load(f)


    # output_dir = "/root/private_data/luog/Data_AbAg/AbAgKer_all/split_fivefold/AbAgIoff_byPDB"
    # with open("/root/private_data/luog/Data_AbAg/AbAgKer_all/json/AbAgAoff_272.json", "r") as f:
    #     data = json.load(f)


    print(f"Total samples: {len(data)}")
    
    folds = split_pdb_data_for_cv(data, n_splits=5, random_seed=42)


    for i, (train_data, val_data) in enumerate(folds):
        print(f"Fold {i+1}: "
            f"Train={len(train_data)}, Val={len(val_data)}, "
            f"Unique PDBs in train: {len(set(d['pdb'][:4] for d in train_data))}, "
            f"Unique PDBs in val: {len(set(d['pdb'][:4] for d in val_data))}")

        # 保存
        with open(os.path.join(output_dir, f"fold_{i+1}_train.json"), "w", encoding="utf-8") as f:
            json.dump(train_data, f, ensure_ascii=False, indent=2)

        with open(os.path.join(output_dir, f"fold_{i+1}_test.json"), "w", encoding="utf-8") as f:
            json.dump(val_data, f, ensure_ascii=False, indent=2)

    print(f"save in: {output_dir}")
```
